#SALE PREDICTION FROM EXISTING CUSTOMNER using LOGISTIC REGRSSION

```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("Digitalad_dataset.csv")
```

```python
print(data.shape)
print(data.head(5))
```

```python
x=data.iloc[:,:-1].values
x
```

```python
y=data.iloc[:,-1:].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(x_train,y_train)
```

```python
age=int(input("Ente the customer's age:"))
salary=int(input("Enter the customer's salary:"))
newcustomer=[[age,salary]]
result=model.predict(sc.transform(newcustomer))
print(result)
if result==1:
    print("The customer will buy")
else:
    print("The customer won't buy")
```

```python
y_pred=model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy of the model is {0}%".format(accuracy_score(y_test,y_pred)*100))
```

```python
#Reciever Operating Curve (ROC Curve)
from sklearn.metrics import roc_auc_score,roc_curve
import matplotlib.pyplot as plt 
nsprob=[0 for _ in range(len(y_test))]
lsprob=model.predict_proba(x_test)
```

```python
#keep probabilities for the positive outcome only
lsprob=lsprob[:,1]
```

```python
#calculate scores
nsAUC=roc_auc_score(y_test,nsprob)
lrAUC=roc_auc_score(y_test,lsprob)
```

```python
#summarize scores
print('NO SKILL:ROC AUC=%.3f'%(nsAUC*100))
print('Logistic skill:ROC AUC=%.3f'%(lrAUC*100))
```

```python
#calculate ROC curves
nsFP,nsTP,_=roc_curve(y_test,nsprob)
lrFP,lrTP,_=roc_curve(y_test,lsprob)
```

```python
#plot thre roc curve
plt.plot(nsFP,nsTP,linestyle='--',label='No Skill')
plt.plot(lrFP,lrTP,marker='*',label='Logistic')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positve Rate')
#show the legend
plt.legend()
plt.show()
```

#cross validation score

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, random_state=None)
result = cross_val_score(model, x,y,cv=kfold)
print("CROSS VALIDATION SCORE: %.2f%%" % (result.mean()*100.0))
```

#stratified K-Fold cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=3,random_state=None)
model_skfold=LogisticRegression()
results_skfold=cross_val_score(model_skfold,x,y,cv=skfold)
print("stratified k-Fold: %.2f%%"%(results_skfold.mean()*100.0))
```

#Cummulative Accuracy profile(CAP) Curve

```python
total=len(y_test)
print(total)
class_1_count=np.sum(y_test)
print(class_1_count)
class_0_count=total-class_1_count
plt.plot([0,total],[0,class_1_count],c='r',linestyle='--',label='Random Model')
plt.plot([0,class_1_count,total],[0,class_1_count,class_1_count],c='grey',linewidth=2,label='Perfet Model')
probs=model.predict_proba(x_test)
probs=probs[:,1]
model_y=[y for _,y in sorted(zip(probs,y_test),reverse=True)]
y_values=np.append([0],np.cumsum(model_y))
x_values=np.arange(0,total+1)
plt.plot(x_values,y_values,c='b',label='LRClassifier',linewidth=4)
index=int((50*total/100))
##50% vertical line from x_axis
plt.plot([index,index],[0,y_values[index]],c='g',linestyle='--')
##Horizontal line to y_axis from prediction model
plt.plot([0,index],[y_values[index],y_values[index]],c='g',linestyle='--')
class_1_observed=y_values[index]*100/max(y_values)
plt.xlabel('Tabel observation')
plt.ylabel('class 1 observation')
plt.title('cummulative accuracy profile')
plt.legend(loc='lower right')
```
